```python
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.io import imread
import numpy as np
import glob
```

```python
idxs = np.asarray([426,223,112,100,31,17,19,500,609,611,713,714,711,701,901,911]) - 1
idxs
```

```python
# classes are indxed from 0
```

```python
plt.figure(figsize=(6,5.5))
for i in range(0, len(idxs)):
    img_name = glob.glob("%i_*.png" % idxs[i])[0]
    #print img_name
    plt.subplot(4,4,i+1)
    if i == 0:
        plt.tight_layout()
    #plt.tight_layout()
    plt.imshow(imread(img_name))
    plt.title(img_name)
    plt.axis('off')
plt.savefig("./plot_faces.png")
```
