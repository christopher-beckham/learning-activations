```python
class DebugNonlinearityLayer(NonlinearityLayer):
    def __init__(self, incoming, nonlinearity=rectify,
                 **kwargs):
        super(DebugNonlinearityLayer, self).__init__(incoming, **kwargs)
        self.prev_layer_input_shape = incoming.input_shape
    def get_output_for(self, input, **kwargs):
        #sys.stderr.write( str(input.shape) + "\n" )
        raise Exception(str(input) + "\n" )
        return self.nonlinearity(input)
```

```python
l_in = InputLayer( (None, 28*28) )
l_dense = DenseLayer(l_in, num_units=20, nonlinearity=linear)
l_act = DebugNonlinearityLayer(l_dense, nonlinearity=rectify)
l_out = DenseLayer(l_act, num_units=10)
```

```python
X = T.fmatrix('X')
```

```python
get_dense = theano.function([X], lasagne.layers.get_output(l_dense, X))
```
